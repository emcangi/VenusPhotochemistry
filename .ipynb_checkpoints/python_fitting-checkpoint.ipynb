```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

# K eddy

```python
keddy = np.loadtxt("../Resources/PlotsFromPapers/keddy_venus_zhang2012.csv")
```

```python
alt=keddy[:, 1]
k = keddy[:, 0]

print(keddy[0:5])
```

```python
plt.scatter(k, alt)
plt.xscale("log")
plt.xlim(1e4, 0.5e7)
plt.show()
```

```python
def func(z, m, b, c):
    return 10**(m*z + b) + c

A = 16
A2 = 11
# do the same fit for irradiance for completeness, even though I don't really need the irradiance data
popt, pcov = curve_fit(func, alt[0:A], k[0:A], maxfev=4000)
popt2, pcov2 = curve_fit(func, alt[A2:], k[A2:], maxfev=4000)

plt.figure(figsize=(10,5))
plt.title("Fitting Keddy")
plt.scatter(k, alt)
plt.plot(func(alt[0:A], popt[0], popt[1], popt[2]), alt[0:A], color="purple", label="fit")
plt.plot(func(alt[A2:], popt2[0], popt2[1], popt2[2]), alt[A2:], color="green", label="fit2")
plt.axhline(alt[A-1])
plt.xlabel("keddy")
plt.xscale("log")
plt.ylabel("alt")
plt.legend()
plt.show()
```
